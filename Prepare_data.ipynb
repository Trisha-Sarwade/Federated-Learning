```python
import pandas as pd
import numpy as np

# Convert date columns to datetime
df2['Post Date - Original'] = pd.to_datetime(df2['Post Date - Original'])

# Define a function to adjust the fiscal year
def adjust_fiscal_year(row):
    fiscal_year_start = pd.to_datetime(f"{row['Fiscal Year']}-07-01")
    if row['Post Date - Original'] < fiscal_year_start:
        return row['Fiscal Year'] - 1
    else:
        return row['Fiscal Year']

# Apply the function to create 'Fiscal Year Adjusted'
df2['Fiscal Year Adjusted'] = df2.apply(adjust_fiscal_year, axis=1)

# Create new fiscal year start date based on the adjusted fiscal year
df2['Fiscal Year Start'] = pd.to_datetime(df2['Fiscal Year Adjusted'].astype(str) + '-07-01')

# Calculate 'contract_term' as the difference in days between 'Post Date - Original' and the adjusted 'Fiscal Year Start'
df2['contract_term'] = (df2['Post Date - Original'] - df2['Fiscal Year Start']).dt.days

# Optionally, check the results
print(df2[['Post Date - Original', 'Fiscal Year', 'Fiscal Year Adjusted', 'Fiscal Year Start', 'contract_term']].head())
```
